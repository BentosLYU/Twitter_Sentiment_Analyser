```python
import matplotlib.pyplot as plt
import os
import pandas as pd
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
```

```python
freq = pd.read_table('../result/li ziqi.txt', sep='\t', header = None)
freq.columns=['Word', 'Frequency']
freq = freq.drop(index = [0,1])
```

```python
ax = freq.plot(xticks = freq.index,kind = 'bar', title = 'Word Frequency', rot = 90)
ax.set_xticklabels(freq.Word)
plt.savefig('../image/frequency.jpg')
plt.show()
```

```python
x = freq["Word"][:20]
y = freq["Frequency"][:20]
plt.barh(x, y)
plt.gca().invert_yaxis()
# plt.xticks(x ,color='black',rotation=-80)
plt.title('Word Frequency', fontsize = 18)
fig = plt.gcf()
plt.xlabel('frequncy', fontsize=18)
plt.ylabel('word', fontsize=18)
# plt.rcParams['xtick.labelsize']=20
plt.rcParams['ytick.labelsize']=18
fig.set_size_inches(8, 6)
fig.savefig('../image/freq.png', dpi=100)
plt.show()
```
